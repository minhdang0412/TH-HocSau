```python
import pandas as pd
import matplotlib.pyplot as plt

# Đọc dữ liệu từ file CSV
df = pd.read_csv("processed_dulieuxettuyendaihoc.csv")
df.head()
# 1. Sắp xếp dữ liệu điểm DH1 theo thứ tự tăng dần
df_sorted_dh1 = df.sort_values(by="DH1")
print(df_sorted_dh1)

# 2. Sắp xếp dữ liệu điểm DH2 tăng dần theo nhóm giới tính
df_sorted_dh2_gt = df.sort_values(by=["GT", "DH2"])
print("Dữ liệu điểm DH2 sau khi sắp xếp tăng dần theo giới tính:")
print(df_sorted_dh2_gt.head())

# 3. Tạo pivot-table thống kê DH1 theo KT
pivot_kt = df.pivot_table(
    values="DH1",
    index="KT",
    aggfunc=[
        "count", "sum", "mean", "median", "min", "max", "std",
        lambda x: x.quantile(0.25),  # Q1
        lambda x: x.quantile(0.5),   # Q2 (median)
        lambda x: x.quantile(0.75)   # Q3
    ]
).rename(columns={
    "<lambda_0>": "Q1",
    "<lambda_1>": "Q2",
    "<lambda_2>": "Q3"
})
print("Pivot-table DH1 theo KT:")
print(pivot_kt)

# 4. Tạo pivot-table thống kê DH1 theo KT và KV
pivot_kt_kv = df.pivot_table(
    values="DH1",
    index=["KT", "KV"],
    aggfunc=["count", "sum", "mean", "median", "min", "max", "std", 
             lambda x: x.quantile(0.25), lambda x: x.quantile(0.5), lambda x: x.quantile(0.75)]
).rename(columns={
    "<lambda_0>": "Q1",
    "<lambda_1>": "Q2",
    "<lambda_2>": "Q3"
})
print("Pivot-table DH1 theo KT và KV:")
print(pivot_kt_kv)

# 5. Tạo pivot-table thống kê DH1 theo KT, KV và DT
pivot_kt_kv_dt = df.pivot_table(
    values="DH1",
    index=["KT", "KV", "DT"],
    aggfunc=["count", "sum", "mean", "median", "min", "max", "std", 
             lambda x: x.quantile(0.25), lambda x: x.quantile(0.5), lambda x: x.quantile(0.75)]
).rename(columns={
    "<lambda_0>": "Q1",
    "<lambda_1>": "Q2",
    "<lambda_2>": "Q3"
})
print("Pivot-table DH1 theo KT, KV và DT:")
print(pivot_kt_kv_dt)
# 6. Lập bảng tần số và tần suất của biến GT
gt_counts = df["GT"].value_counts()
gt_frequencies = df["GT"].value_counts(normalize=True) * 100
print("Bảng tần số GT:")
print(gt_counts)
print("Bảng tần suất GT (%):")
print(gt_frequencies)

# 7. Vẽ biểu đồ tần số (cột) và tần suất (tròn) của GT
plt.figure(figsize=(10, 5))
gt_counts.plot(kind='bar', color='skyblue', title='Biểu đồ tần số GT')
plt.xlabel('Giới tính')
plt.ylabel('Tần số')
plt.show()

plt.figure(figsize=(6, 6))
gt_frequencies.plot(kind='pie', autopct='%1.1f%%', colors=['lightcoral', 'lightskyblue'], title='Biểu đồ tần suất GT')
plt.ylabel('')
plt.show()

# 8. Trình bày dữ liệu lần lượt của US_TBM1, US_TBM2, US_TBM3
print("Mô tả dữ liệu US_TBM1:")
print(df["US_TBM1"].describe())
print("Mô tả dữ liệu US_TBM2:")
print(df["US_TBM2"].describe())
print("Mô tả dữ liệu US_TBM3:")
print(df["US_TBM3"].describe())

# 9. Trình bày dữ liệu biến DT với học sinh là nam
dt_nam = df[df["GT"] == "Nam"]["DT"]
print("Dữ liệu DT của học sinh nam:")
print(dt_nam.describe())

# 10. Trình bày dữ liệu biến KV với học sinh nam thuộc dân tộc Kinh và có điểm đạt yêu cầu
kv_nam_kinh = df[(df["GT"] == "Nam") & (df["DT"] == "Kinh") & (df["DH1"] >= 5.0) & (df["DH2"] >= 4.0) & (df["DH3"] >= 4.0)]["KV"]
print("Dữ liệu KV của học sinh nam dân tộc Kinh đạt điểm yêu cầu:")
print(kv_nam_kinh.describe())

# 11. Trình bày dữ liệu DH1, DH2, DH3 lớn hơn hoặc bằng 5.0 và thuộc khu vực 2NT
dh_5_kv2nt = df[(df["DH1"] >= 5.0) & (df["DH2"] >= 5.0) & (df["DH3"] >= 5.0) & (df["KV"] == "2NT")]
print("Dữ liệu DH1, DH2, DH3 >= 5.0 và thuộc khu vực 2NT:")
print(dh_5_kv2nt.describe())

#12 
# Lọc dữ liệu học sinh nữ (GT = "F")
df_female = df[df["GT"] == "F"]

# Đếm số lượng học sinh theo xếp loại trong từng nhóm XL1, XL2, XL3
xl_counts = {
    "XL1": df_female["XL1"].value_counts(),
    "XL2": df_female["XL2"].value_counts(),
    "XL3": df_female["XL3"].value_counts(),
}

# Chuyển đổi dữ liệu thành DataFrame
df_counts = pd.DataFrame(xl_counts).fillna(0)

# Cập nhật thứ tự xếp loại phù hợp với dữ liệu thực tế
xl_order = ["Y", "TB", "K", "G"]  # Loại bỏ 'XS' vì không có trong dữ liệu
df_counts = df_counts.reindex(xl_order)

# Vẽ biểu đồ bằng Matplotlib
fig, ax = plt.subplots(figsize=(10, 6))
bar_width = 0.2
x = range(len(df_counts.index))

# Vẽ từng nhóm xếp loại
for i, col in enumerate(df_counts.columns):
    ax.bar(
        [pos + i * bar_width for pos in x],  # Dịch vị trí cột
        df_counts[col], 
        width=bar_width, 
        label=col
    )

# Thiết lập nhãn và tiêu đề
ax.set_title("Phân bố học sinh nữ theo nhóm xếp loại", fontsize=14)
ax.set_xlabel("Xếp loại", fontsize=12)
ax.set_ylabel("Số lượng học sinh", fontsize=12)
ax.set_xticks([pos + bar_width for pos in x])
ax.set_xticklabels(df_counts.index, rotation=0)
ax.legend(title="Nhóm XL")

# Hiển thị biểu đồ
plt.show()

#13
# Lọc dữ liệu học sinh có khối thi A, A1, B và thuộc khu vực 1, 2
df_filtered = df[(df["KT"].isin(["A", "A1", "B"])) & (df["KV"].isin(["1", "2"]))]

# Thống kê số lượng học sinh theo kết quả xét tuyển trong từng khối thi
kqxt_counts = df_filtered.groupby(["KT", "KQXT"]).size().unstack(fill_value=0)

# Vẽ biểu đồ bằng Matplotlib
fig, ax = plt.subplots(figsize=(10, 6))
bar_width = 0.3
x = range(len(kqxt_counts.index))

# Màu sắc đại diện cho KQXT (0: Không trúng tuyển, 1: Trúng tuyển)
colors = ["red", "green"]

# Vẽ từng nhóm kết quả xét tuyển
for i, col in enumerate(kqxt_counts.columns):
    ax.bar(
        [pos + i * bar_width for pos in x],  # Dịch vị trí cột
        kqxt_counts[col], 
        width=bar_width, 
        label=f"KQXT {int(col)}", 
        color=colors[i]
    )

# Thiết lập nhãn và tiêu đề
ax.set_title("Phân bố kết quả xét tuyển theo khối thi", fontsize=14)
ax.set_xlabel("Khối thi", fontsize=12)
ax.set_ylabel("Số lượng học sinh", fontsize=12)
ax.set_xticks([pos + bar_width / 2 for pos in x])
ax.set_xticklabels(kqxt_counts.index, rotation=0)
ax.legend(title="Kết quả xét tuyển")

# Hiển thị biểu đồ
plt.show()

#14
# Nhóm dữ liệu theo khu vực (KV) và khối thi (KT), đếm số lượng thí sinh
kv_khoi_counts = df.groupby(["KV", "KT"]).size().unstack(fill_value=0)

# Vẽ biểu đồ bằng Matplotlib
fig, ax = plt.subplots(figsize=(12, 6))
bar_width = 0.15
x = range(len(kv_khoi_counts.index))

# Vẽ từng nhóm khối thi
for i, col in enumerate(kv_khoi_counts.columns):
    ax.bar(
        [pos + i * bar_width for pos in x],  # Dịch vị trí cột
        kv_khoi_counts[col], 
        width=bar_width, 
        label=f"Khối {col}"
    )

# Thiết lập nhãn và tiêu đề
ax.set_title("Số lượng thí sinh từng khu vực theo khối thi", fontsize=14)
ax.set_xlabel("Khu vực", fontsize=12)
ax.set_ylabel("Số lượng thí sinh", fontsize=12)
ax.set_xticks([pos + (len(kv_khoi_counts.columns) / 2) * bar_width for pos in x])
ax.set_xticklabels(kv_khoi_counts.index, rotation=0)
ax.legend(title="Khối thi")

# Hiển thị biểu đồ
plt.show()

# Nhóm dữ liệu theo khối thi (KT) và kết quả xét tuyển (KQXT), đếm số lượng thí sinh
khoi_kqxt_counts = df.groupby(["KT", "KQXT"]).size().unstack(fill_value=0)

# Vẽ biểu đồ số lượng thí sinh đậu, rớt theo từng khối thi
fig, ax = plt.subplots(figsize=(10, 6))
bar_width = 0.3
x = range(len(khoi_kqxt_counts.index))

colors = ["red", "green"]

for i, col in enumerate(khoi_kqxt_counts.columns):
    ax.bar(
        [pos + i * bar_width for pos in x],
        khoi_kqxt_counts[col], 
        width=bar_width, 
        label=f"KQXT {int(col)}", 
        color=colors[i]
    )

ax.set_title("Số lượng thí sinh đậu, rớt theo từng khối thi", fontsize=14)
ax.set_xlabel("Khối thi", fontsize=12)
ax.set_ylabel("Số lượng thí sinh", fontsize=12)
ax.set_xticks([pos + bar_width / 2 for pos in x])
ax.set_xticklabels(khoi_kqxt_counts.index, rotation=0)
ax.legend(title="Kết quả xét tuyển")
plt.show()

#15

# Nhóm dữ liệu theo khu vực (KV) và kết quả xét tuyển (KQXT), đếm số lượng thí sinh
kv_kqxt_counts = df.groupby(["KV", "KQXT"]).size().unstack(fill_value=0)

# Vẽ biểu đồ số lượng thí sinh đậu, rớt theo khu vực
fig, ax = plt.subplots(figsize=(10, 6))
bar_width = 0.3
x = range(len(kv_kqxt_counts.index))

colors = ["red", "green"]

for i, col in enumerate(kv_kqxt_counts.columns):
    ax.bar(
        [pos + i * bar_width for pos in x],
        kv_kqxt_counts[col], 
        width=bar_width, 
        label=f"KQXT {int(col)}", 
        color=colors[i]
    )

ax.set_title("Số lượng thí sinh đậu, rớt theo khu vực", fontsize=14)
ax.set_xlabel("Khu vực", fontsize=12)
ax.set_ylabel("Số lượng thí sinh", fontsize=12)
ax.set_xticks([pos + bar_width / 2 for pos in x])
ax.set_xticklabels(kv_kqxt_counts.index, rotation=0)
ax.legend(title="Kết quả xét tuyển")
plt.show()
#Nhóm dữ liệu theo dân tộc (DT) và kết quả xét tuyển (KQXT), đếm số lượng thí sinh
dt_kqxt_counts = df.groupby(["DT", "KQXT"]).size().unstack(fill_value=0)

# Vẽ biểu đồ số lượng thí sinh đậu, rớt theo dân tộc
fig, ax = plt.subplots(figsize=(12, 6))
bar_width = 0.3
x = range(len(dt_kqxt_counts.index))

colors = ["red", "green"]

for i, col in enumerate(dt_kqxt_counts.columns):
    ax.bar(
        [pos + i * bar_width for pos in x],
        dt_kqxt_counts[col], 
        width=bar_width, 
        label=f"KQXT {int(col)}", 
        color=colors[i]
    )

ax.set_title("Số lượng thí sinh đậu, rớt theo nhóm dân tộc", fontsize=14)
ax.set_xlabel("Dân tộc", fontsize=12)
ax.set_ylabel("Số lượng thí sinh", fontsize=12)
ax.set_xticks([pos + bar_width / 2 for pos in x])
ax.set_xticklabels(dt_kqxt_counts.index, rotation=90)
ax.legend(title="Kết quả xét tuyển")
plt.show()

# Nhóm dữ liệu theo giới tính (GT) và kết quả xét tuyển (KQXT), đếm số lượng thí sinh
gt_kqxt_counts = df.groupby(["GT", "KQXT"]).size().unstack(fill_value=0)

# Vẽ biểu đồ số lượng thí sinh đậu, rớt theo giới tính
fig, ax = plt.subplots(figsize=(8, 5))
bar_width = 0.3
x = range(len(gt_kqxt_counts.index))

colors = ["red", "green"]

for i, col in enumerate(gt_kqxt_counts.columns):
    ax.bar(
        [pos + i * bar_width for pos in x],
        gt_kqxt_counts[col], 
        width=bar_width, 
        label=f"KQXT {int(col)}", 
        color=colors[i]
    )

ax.set_title("Số lượng thí sinh đậu, rớt theo giới tính", fontsize=14)
ax.set_xlabel("Giới tính", fontsize=12)
ax.set_ylabel("Số lượng thí sinh", fontsize=12)
ax.set_xticks([pos + bar_width / 2 for pos in x])
ax.set_xticklabels(gt_kqxt_counts.index, rotation=0)
ax.legend(title="Kết quả xét tuyển")
plt.show()





```
